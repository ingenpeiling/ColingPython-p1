```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('administration_data.csv')
```

```python
index_names = data[data['gender'] == "Male"].index
female = data.drop(index_names)

index_names = data[data['gender'] == "Female"].index
male = data.drop(index_names)
```

```python
plt.figure(figsize=(25, 15))

sns.lineplot(x = "age", y = "comprehension", data = male, label='male comprehension')
sns.lineplot(x = "age", y = "production", data = male, label='male production')
sns.lineplot(x = "age", y = "comprehension", data = female, label='female comprehension')
sns.lineplot(x = "age", y = "production", data = female, label='female production')

plt.xticks(range(6, 37, 6), fontsize=14)
plt.yticks(range(0, 701, 100), fontsize=14)

plt.grid()

plt.xlabel('Возраст детей в месяцах', fontsize=21)
plt.ylabel('Количество слов', fontsize=21)

plt.legend(fontsize=21)
plt.title('Динамика усвоения русского языка как родного', fontsize=36)
```

"Слияние" восприятия и порождения слов наступает в возрасте около 19-20 месяцев (согласно приведенным данным).
